**This program stores two numbers in variables, adds them together, and displays the result.**


```python
a = 10
b = 20

sum = a + b

print("Sum =", sum)
```

**This program checks if the remainder after dividing the number by 2 is zero. If yes, it prints Even; otherwise, it prints Odd.**

```python
num = 18

if num % 2 == 0:
    print(num, "is Even")
else:
    print(num, "is Odd")
```

**This program stores five student names in a list and uses a for loop to display each name.**

```python
students = ["Rahul", "Priya", "Amit", "Neha", "John"]

print("Student Names:")

for student in students:
    print(student)
```

**This program compares three numbers using conditional statements and prints the largest value.**

```python
a = 25
b = 40
c = 18

largest = max(a, b, c)

print("Largest number:", largest)
```

**This program uses a for loop to multiply numbers from 1 to the given number.**

```python
number = 5
factorial = 1

for i in range(1, number + 1):
    factorial *= i

print("Factorial:", factorial)
```

**This program uses string slicing to reverse the characters in a string.**

```python
text = "Python"

reverse = text[::-1]

print("Reversed String:", reverse)
```

**This program uses the min() function to find the smallest element.**

```python
numbers = [45, 12, 89, 23, 7]

smallest = min(numbers)

print("Smallest Number:", smallest)
```

**This program uses the max() function to find the largest element.**

```python
numbers = [45, 12, 89, 23, 7]

largest = max(numbers)

print("Largest Number:", largest)
```

**This program uses the len() function to count the elements.**
