**Create an Employee table, insert sample records, and display all records..**


```python
import sqlite3

# Create an in-memory SQLite database
connection = sqlite3.connect(":memory:")
cursor = connection.cursor()

# Create Employee table
cursor.execute("""
CREATE TABLE Employee (
    EmpID INTEGER PRIMARY KEY,
    Name TEXT,
    Department TEXT,
    Salary INTEGER
)
""")

# Insert sample records
cursor.executemany("""
INSERT INTO Employee (EmpID, Name, Department, Salary)
VALUES (?, ?, ?, ?)
""", [
    (101, "Rahul", "IT", 50000),
    (102, "Priya", "HR", 45000),
    (103, "Amit", "Sales", 55000)
])

# Display all records
cursor.execute("SELECT * FROM Employee")
records = cursor.fetchall()

print("Employee Records:")
for record in records:
    print(record)

# Close the connection
connection.close()
```

**Write a Python program using SQLite to retrieve employees whose salary is greater than 50000.**

```python
import sqlite3

# Create an in-memory database
connection = sqlite3.connect(":memory:")
cursor = connection.cursor()

# Create Employee table
cursor.execute("""
CREATE TABLE Employee (
    EmpID INTEGER PRIMARY KEY,
    Name TEXT,
    Department TEXT,
    Salary INTEGER
)
""")

# Insert sample data
cursor.executemany("""
INSERT INTO Employee VALUES (?, ?, ?, ?)
""", [
    (101, "Rahul", "IT", 50000),
    (102, "Priya", "HR", 45000),
    (103, "Amit", "Sales", 55000),
    (104, "Neha", "IT", 60000)
])

# SQL query
cursor.execute("""
SELECT * FROM Employee
WHERE Salary > 50000
""")

rows = cursor.fetchall()

print("Employees with Salary > 50000:")

for row in rows:
    print(row)

connection.close()
```

**Write a Python program using SQLite to update the salary of an employee.**

```python
import sqlite3

# Create an in-memory database
connection = sqlite3.connect(":memory:")
cursor = connection.cursor()

# Create Employee table
cursor.execute("""
CREATE TABLE Employee (
    EmpID INTEGER PRIMARY KEY,
    Name TEXT,
    Department TEXT,
    Salary INTEGER
)
""")

# Insert sample records
cursor.executemany("""
INSERT INTO Employee VALUES (?, ?, ?, ?)
""", [
    (101, "Rahul", "IT", 50000),
    (102, "Priya", "HR", 45000),
    (103, "Amit", "Sales", 55000)
])

# Update salary
cursor.execute("""
UPDATE Employee
SET Salary = 60000
WHERE EmpID = 101
""")

# Display updated records
cursor.execute("SELECT * FROM Employee")

rows = cursor.fetchall()

print("Updated Employee Records:")

for row in rows:
    print(row)

connection.commit()
connection.close()
```

**Write a Python program using SQLite to delete an employee record from the Employee table.**

```python
import sqlite3

connection = sqlite3.connect(":memory:")
cursor = connection.cursor()

cursor.execute("""
CREATE TABLE Employee (
    EmpID INTEGER PRIMARY KEY,
    Name TEXT,
    Department TEXT,
    Salary INTEGER
)
""")

cursor.executemany("""
INSERT INTO Employee VALUES (?, ?, ?, ?)
""", [
    (101, "Rahul", "IT", 50000),
    (102, "Priya", "HR", 45000),
    (103, "Amit", "Sales", 55000)
])

# Delete one record
cursor.execute("""
DELETE FROM Employee
WHERE EmpID = 102
""")

# Display remaining records
cursor.execute("SELECT * FROM Employee")

rows = cursor.fetchall()

print("Remaining Employee Records:")

for row in rows:
    print(row)

connection.commit()
connection.close()
```

**Write a Python program using SQLite to display employee records in ascending order of salary.**

```python
import sqlite3

connection = sqlite3.connect(":memory:")
cursor = connection.cursor()

cursor.execute("""
CREATE TABLE Employee (
    EmpID INTEGER PRIMARY KEY,
    Name TEXT,
    Department TEXT,
    Salary INTEGER
)
""")

cursor.executemany("""
INSERT INTO Employee VALUES (?, ?, ?, ?)
""", [
    (101, "Rahul", "IT", 50000),
    (102, "Priya", "HR", 45000),
    (103, "Amit", "Sales", 55000),
    (104, "Neha", "IT", 60000)
])

cursor.execute("""
SELECT * FROM Employee
ORDER BY Salary ASC
""")

rows = cursor.fetchall()

print("Employees Sorted by Salary:")

for row in rows:
    print(row)

connection.close()
```
